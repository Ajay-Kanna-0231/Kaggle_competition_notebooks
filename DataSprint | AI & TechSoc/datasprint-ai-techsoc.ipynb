```python
from scipy.sparse import csr_matrix
import pandas as pd

# Assuming numerical_vectors is a sparse matrix
# Calculate the mean ignoring zeros (i.e., only non-zero entries contribute to the mean)

# row_means is a 2D matrix of shape (n_samples, 1). We need to flatten it to 1D.
row_means_2 = np.ravel(row_means_2)  # This flattens the 2D array to 1D

# Now, check the length of row_means and your DataFrame's transcription column after dropping NA
print(len(row_means_2))  # This should now match the number of non-null entries in `transcription`

# Assuming train_data is your DataFrame and it's indexed sequentially without gaps
# If there are nulls in `transcription`, we should align the indices first
non_null_transcriptions_index_2 = train_data.dropna(subset=['sample_name']).index

# Ensure the DataFrame index matches the row_means calculation
train_data.loc[non_null_transcriptions_index_2, 'row_means_2'] = row_means_2

# If there were any discrepancies in lengths, this step ensures that we're aligning the data correctly
print(train_data)
```

```python
new_set = train_data[['medical_specialty', 'row_means', 'row_means_2']]
print(new_set)
```

```python
new_set = new_set.dropna()
print(new_set)
new_set_encoded = pd.get_dummies(new_set, columns=['medical_specialty'], prefix='specialty')

new_set_encoded = new_set_encoded.astype(float)
print(new_set_encoded)
X = new_set_encoded[['row_means', 'row_means_2']]
print(X)
y = new_set_encoded.drop(columns=X.columns)
print(y)

#y = new_set_encoded[]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.9, test_size=0.1, random_state=0)
print(X_train)
print(X_valid)
print(y_train)
print(y_valid)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=500, learning_rate = 0.1)
model.fit(X_train,y_train)
prediction = model.predict(X_valid)
print(prediction)
print(prediction.shape)
```

```python
y_train.columns = y_train.columns.str.replace('specialty_', '')
column_headers = list(y_train.columns)
prediction_df = pd.DataFrame(prediction, columns=column_headers)
pred = []
num = 0
for i in prediction:
    num += 1
    max = float('-inf')
    col_no = -1
    for j in i:
        col_no += 1
        if float(j) > max:
            max = float(j)
            col = col_no
    pred_name = prediction_df.columns[col]
    pred.append(pred_name)   

print(pred)
```

```python
acc = 0
for i in pred:
    print(i)
    if y_valid.iloc[i] == 1:
        acc += 1
print(acc)
```
